```python
import pandas as pd
import re
```

```python
df1=pd.read_csv('flipkart_product.csv', encoding='latin-1')
df=df1.copy()
```

```python
print(df.shape)
print(df.columns)
print(df.head())
```

```python
df['ProductName']=df['ProductName'].apply(lambda x: re.sub(r'[^\x00-\x7F]','',str(x)).split())
```

```python
df
```

```python
df['Price']=df['Price'].astype(str).apply(lambda x: re.sub(r'[^\d]','', x))
df['Price']=pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['Rate']= pd.to_numeric(df['Rate'], errors='coerce')
```

```python
df.dropna(subset=['Review', 'Rate'], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
print(df.shape)
print(df.dtypes)
print(df.head())
```

```python
df['ProductName']=df['ProductName'].apply(lambda x:' '.join(x) if isinstance(x, list) else str(x))
print(df['ProductName'].head())
```

```python
print(type(df['ProductName'][0]))
print(repr(df['ProductName'][0]))
```

```python
df['ProductName']= df['ProductName'].apply(lambda x: re.sub(r'\?+', '', str(x)))
```

```python
df['ProductName']
```

```python
df
```

```python
def label_sentiment(rate):
  if rate>=4.0:
    return 'Positive'
  elif rate==3:
    return 'Neutral'
  else:
    return 'Negative'
df['Sentiment']=df['Rate'].apply(label_sentiment)


```

```python
print(df['Sentiment'].value_counts())
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
df['compound_score']=df['Review'].astype(str).apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
def vader_label(score):
  if score >= 0.05:
    return 'Positive'
  elif score <= -0.05:
    return 'Negative'
  else:
    return 'Neutral'
df['VADER_Sentiment']= df['compound_score'].apply(vader_label)

print("VADER Sentiment Distribution:")
print(df['VADER_Sentiment'].value_counts())
```

```python
print(df[['Review', 'Sentiment', 'VADER_Sentiment']].head(10))
```

```python
print("VADER Sentiment Distribution:")
print(df['VADER_Sentiment'].value_counts())
print("\n Star Rating Sentiment Distribution:")
print(df['Sentiment'].value_counts())
```

```python
import sqlite3
conn=sqlite3.connect('flipkart_reviews.db')
df.to_sql('reviews', conn, if_exists='replace', index=False)
print("\n Total rows in database:")
print(pd.read_sql("Select Count(*) as total_rows From reviews", conn))
```

Sentiment Distribution

```python
q1= pd.read_sql(""" Select Sentiment, COUNT(*) as total
From reviews group by sentiment order by total desc""", conn)
print("Sentiment Distribution:")
print(q1)
```

Top 10 most reviewed products

```python
q2=pd.read_sql("""Select ProductName, Count(*) as review_count,
Round(avg(Rate),2) as avg_rating from reviews
Group by ProductName order by review_count DESC LIMIT 10 """,conn)
print("Top 10 Most Reviewed Products:")
print(q2)
```
